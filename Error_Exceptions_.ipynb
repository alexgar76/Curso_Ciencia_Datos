```python
#example 1 To handle exception python use a TRY statement
try:
  1/0
except:
  print("Zero divison error")
```

```python
#example 2 to only handle a certain exception  we can write the exception after the EXCEPT
try:
  1/0
except ZeroDivisionError:
  print("Zero divison error")
```

```python
#example 3 to handle more that one exception
try:
  1/0 # Throws a ZeroDivisionError exception
  print(Undefined_variable) #Throws a Name Error Exception

except (ZeroDivisionError,NameError):
  print("Zero divison error")
```

```python
#example 4 Assertion Error El uso de assert en Python nos permite realizar comprobaciones. Si la expresión contenida dentro del mismo es False, se lanzará una excepción,
#concretamente AssertionError.
assert 1==1
```
