```python
#ejm 3 (Indice de paquetes Python PyPI)
#Libreria Matplotlib(libreria de grafico)
import matplotlib.pyplot as plt

# X Representa los años
x=[2012,2013,2014,2015,2016,2017]

#Y representa el dinero ahorrado
y=[0,50,70,70,100,246]

plt.plot(x,y)
plt.xlabel("Años")
plt.ylabel("Dinero ahorrado")
plt.title("Dinero ahorrado por año")
plt.show()
```

```python
#ejm 4 Grafica de barras con matplotlib
import matplotlib.pyplot as plt

# X representa los años
x=[2017,2018,2019,2020,2021]

#Y representa los terremotos
y=[1566,1808,1637,1433,2206]

plt.bar(x,y)
plt.xlabel("Año")
plt.ylabel("terrenos")
plt.title("Terremoto por año")
plt.show()
```
