```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
#suma, resta, multiplicacion y division de arrays
sum_arr = arr1 + arr2
diff_arr = arr1 - arr2
mult_arr = arr1 * arr2
div_arr =arr1 / arr2
print(sum_arr)
print(diff_arr)
print(mult_arr)
print(div_arr)
```

```python
#funciones universale (ufuncs)


arr = np.array([0,np.pi/2,np.pi])
print(arr)
sin_arr = np.sin(arr)
cos_arr = np.cos(arr)
exp_arr = np.exp(arr)
log_arr = np.log(arr+1) #evitar log (0) que serultaria en  -inf
tan_arr = np.tan(arr)

print(sin_arr)
print(cos_arr)
print(exp_arr)
print(log_arr)
print(tan_arr)
```

```python
#operaciones de algebra lineal
matrix1 = np.array([[1,2],[3,4]])
matrix2 = np.array([[5,6],[7,8]])
#multiplicacion de matrices
mat_mult = np.dot(matrix1,matrix2)
print(mat_mult)
```
